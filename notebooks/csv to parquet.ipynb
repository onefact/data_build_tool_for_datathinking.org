```python
#have your file loaded into a pandas DataFrame using the matching 'read' function

#df = pd.read_json('your_dataset.csv')
#df = pd.read_excel('your_dataset.csv')

df = pd.read_csv('your_dataset.csv')


#check that the dataframe is formatted correctly
df.head()
```

```python
#specify where you want to download this file as a parquet with your path,
#as well as what you want to name your file

df.to_parquet('destination_path/your_dataset.parquet')
```

```python
#load your new parquet file as a pandas dataframe to check that it worked

df_check = pd.read_parquet('your_dataset.parquet')

df_check.head()
```
